To visualize pour model predictions lets make function :

* take in trained model, X and y
* create meshgrid of different X values
* Make Predictions across meshgrid
* plot predictions as well as a line between the zones

```python
def plot_decision_boudaries(model, X, y):
  x_min, x_max = X[:, 0].min(), X[:,0].max()
  y_min, y_max = X[:, 1].min(), X[:,1].max()
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))
  x_in = np.c_[xx.ravel(), yy.ravel()]
  y_pred = model.predict(x_in)

  if len(y_pred[0])>1:
    y_pred = np.argmax(y_pred, axis=1).reshape=(xx.shape)
  else:
    y_pred = np.round(y_pred).reshape(xx.shape)
  
  plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
```

```python
plot_decision_boudaries(model, X, y)
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
  tf.keras.layers.Dense(5, activation='relu'),
  tf.keras.layers.Dense(1)
])

model.compile(loss = tf.keras.losses.BinaryCrossentropy(),
              optimizer = tf.keras.optimizers.Adam(),
              metrics = ['accuracy'])

model.fit(X, y, epochs=100)
```

```python
plot_decision_boudaries(model, X, y)
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
  tf.keras.layers.Dense(10, activation='relu'),
  tf.keras.layers.Dense(5, activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss = tf.keras.losses.BinaryCrossentropy(),
              optimizer = tf.keras.optimizers.Adam(),
              metrics = ['accuracy'])

model.fit(X, y, epochs=100, verbose=0)
model.evaluate(X, y)
```

```python
plot_decision_boudaries(model, X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
  tf.keras.layers.Dense(10, activation='relu'),
  tf.keras.layers.Dense(5, activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss = tf.keras.losses.BinaryCrossentropy(),
              optimizer = tf.keras.optimizers.Adam(),
              metrics = ['accuracy'])

model.fit(X_train, y_train, epochs=100, verbose=0)
model.evaluate(X_test, y_test)
```

```python
plot_decision_boudaries(model, X_train, y_train)
```
